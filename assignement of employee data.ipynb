```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Employee data.csv')

salaries = df['salary']

plt.hist(salaries, bins=20, color='skyblue', edgecolor='black')

plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Employee data.csv')

average_salary_by_gender = df.groupby('gender')['salary'].mean()

average_salary_by_gender.plot(kind='bar', color=['skyblue', 'purple'], edgecolor='black')

plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.title('Average Salary by Gender')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Employee data.csv')

prevexp = df['prevexp']
salary = df['salary']

plt.scatter(prevexp, salary, color='green', edgecolors='black', alpha=0.7)

plt.xlabel('Previous Work Experience')
plt.ylabel('Current Salary')
plt.title('Relationship between Previous Work Experience and Current Salary')

plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Employee data.csv')

education_counts = df['educ'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(education_counts, labels=None, autopct='%1.1f%%', startangle=90, pctdistance=0.85, labeldistance=1.1)

plt.legend(education_counts.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Distribution of Educational Backgrounds among Employees')

plt.axis('equal')

plt.show()
```
